### Data engineering



Après avoir chargé Le jeu de données du fichier `./Data/Heights/Galton.txt` :

1.  vérifier s'il y a des données manquantes ou aberrantes et les corriger si besoin.



```python
# TODO
import pandas as pd

# Charger le jeu de données
df = pd.read_csv('./Data/Heights/Galton.txt', delimiter='\t')

# Vérifier les données manquantes
missing_data = df.isna().sum()
print("Données manquantes :\n", missing_data)

# Examiner les statistiques descriptives
descriptive_stats = df.describe()
print("Statistiques descriptives :\n", descriptive_stats)

# Corriger les données manquantes en remplaçant par la moyenne de la colonne respective
# Par exemple, remplacez les valeurs manquantes dans la colonne 'Height' par la moyenne :
df['Height'].fillna(df['Height'].mean(), inplace=True)

# Corriger les valeurs aberrantes en supprimant les lignes en dehors du seuil basé sur l'écart type
threshold = 3  # Définir un seuil en fonction de la déviation standard
df = df[(df['Height'] - df['Height'].mean()).abs() < threshold * df['Height'].std()]
```

1.  convertir les tailles de pouces en centimètres



```python
# TODO
# Convertir les tailles de pouces en centimètres
df['Height'] = df['Height'] * 2.54
```

1.  créer un nouvel attribut `Average` qui sera la moyenne des tailles du père et de la mère .


```python
# TODO
# Créer un nouvel attribut 'Average' qui est la moyenne des tailles du père et de la mère
df['Average'] = (df['Father'] + df['Mother']) / 2
```

### Moyennes



1.  Sélectionnez les familles où `Average` est dans le plus grand quartile. Quelle est la taille moyenne des parents et celle des enfants dans ce quartile ? Sont-elles *significativement* différentes ?



```python
# Calculer les quantiles pour déterminer le seuil du quartile supérieur
quartile_threshold = df['Average'].quantile(0.75)

# Sélectionner les familles où Average est dans le plus grand quartile
selected_families = df[df['Average'] > quartile_threshold]

# Calculer la taille moyenne des parents dans ce quartile
mean_parent_height = selected_families[['Father', 'Mother']].mean().mean()

# Calculer la taille moyenne des enfants dans ce quartile
mean_child_height = selected_families['Height'].mean()

# Afficher les résultats
print("Taille moyenne des parents dans le quartile supérieur:", mean_parent_height)
print("Taille moyenne des enfants dans le quartile supérieur:", mean_child_height)
```

1.  Sélectionnez les familles où `Average` est dans le plus petit quartile. Quelle est la taille moyenne des parents et celle des enfants dans ce quartile ? Sont-elles *significativement* différentes ?


```python
# Calculer les quantiles pour déterminer le seuil du quartile inférieur
quartile_threshold = df['Average'].quantile(0.25)

# Sélectionner les familles où Average est dans le plus petit quartile
selected_families = df[df['Average'] < quartile_threshold]

# Calculer la taille moyenne des parents dans ce quartile
mean_parent_height = selected_families[['Father', 'Mother']].mean().mean()

# Calculer la taille moyenne des enfants dans ce quartile
mean_child_height = selected_families['Height'].mean()

# Afficher les résultats
print("Taille moyenne des parents dans le quartile inférieur:", mean_parent_height)
print("Taille moyenne des enfants dans le quartile inférieur:", mean_child_height)
```
