Calculer les taux d'acceptation pour les femmes et pour les hommes
dans le jeu de données du fichier
`./Data/Simpson-Paradox/UC-Berkeley-applicants-no-dept.csv`.

-   Lecture des données:


```python
#%load hints/exo_descr-stats_11-1.py
#TODO
simpson = pd.read_csv("./Data/Simpson-Paradox/UC-Berkeley-applicants-no-dept.csv")
simpson.info()
```

-   Calcul des taux d'acceptation :


```python
#%load hints/exo_descr-stats_12-1.py
#TODO
admit = simpson.loc[:,['Admit']]
admitgrouped = admit.groupby('Admit')
rej = 0
adm = 0
for cle, val in admitgrouped:
    if cle == 'Admitted':
        adm = val.count()
    else:
        rej = val.count()
tauxadmis = (adm*100/admit.count()).iloc[0]
tauxrej = (rej*100/admit.count()).iloc[0]
print('taux d admis = ', tauxadmis)
print('taux de rejette = ',tauxrej)
```
